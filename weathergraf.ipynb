```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Загрузка данных о погоде из файла JSON

```python
with open('weather_data.json', 'r', encoding='utf-8') as f:
  weather_data = json.load(f)
```

# Создание DataFrame из загруженных данных

```python
df = pd.DataFrame.from_dict(weather_data, orient='index')
```

# График изменения температуры в разных городах

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=df.index, y='temperature', data=df)
plt.title('Изменение температуры в разных городах')
plt.xlabel('Города')
plt.ylabel('Температура (°C)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# График распределения температуры

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['temperature'], bins=10, kde=True)
plt.title('Распределение температуры')
plt.xlabel('Температура (°C)')
plt.ylabel('Частота')
plt.tight_layout()
plt.show()
```
